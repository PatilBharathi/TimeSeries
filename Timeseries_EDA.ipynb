Lets findout what percentage of data has missing values

```python
# Function to create a data frame with number and percentage of missing data in a data frame

def missingdata_percentage(data):

  total_missing_count = data.isnull().sum().sort_values(ascending = False)
  percent_missing_count = round((data.isnull().sum() / data.isnull().count()*100).sort_values(ascending = False),2)
  missing_data = pd.concat([total_missing_count, percent_missing_count], axis = 1, keys=['Total_count' , 'percent'])
  return missing_data
```

```python
missing_df = missingdata_percentage(df)
missing_df[missing_df['Total_count'] > 0]
```

We can remove/replace the outliers using many techniques.

==> IQR

==> Z score, standard deviation and so on.


But it is not always recomended because it may add some insights to data.

For time being, let us copy the dataframe and remove outliers so that it doesnt impact the original data

```python
def remove_outliers(df,col):
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  ul = q3 + 1.5*iqr
  ll = q1 - 1.5*iqr
  return df[(df[col] > ll) & (df[col] < ul)]
```

```python
df_PM2 = remove_outliers(df,"PM2.5")
```

```python
sns.boxplot(data=df_PM2, x="PM2.5")
plt.show()
```
